```python
#read CSV data

import pandas as pd
import os

```

```python
#read the excel file

crowdfunding_data=pd.ExcelFile("crowdfunding.xlsx")

#get the sheets name


crowdfunding_data.sheet_names

# get the crowdfunding_info from the worksheet

crowdfunding_info_df = pd.read_excel(crowdfunding_data, sheet_name= 'crowdfunding_info')

crowdfunding_info_df.head()

#get the contact_info from the worksheet
# increase width of the column to 400

pd.set_option('max_colwidth', 400)

contact_info_df=pd.read_excel(crowdfunding_data, sheet_name='contact_info',header=3)

contact_info_df.head()
```

```python
# get the unique values in category & subcategory column.

crowdfunding_info_df['category & sub-category'].unique()

# count number of unique values 2 ways:

x =len(pd.unique(crowdfunding_info_df['category & sub-category']))
x

# variable to hold the count
cnt = 0

hello=[]
# loop for counting the unique
# values in category & subcategory
for i in range(0, len(crowdfunding_info_df['category & sub-category'])):
    
    if crowdfunding_info_df['category & sub-category'][i] not in hello: 
        
        hello.append(crowdfunding_info_df['category & sub-category'][i])
          
        cnt += 1
  
print("No.of.unique values :",
      cnt)
```

```python
# split the category and subcategory column using .str.split
# the '/' splits at the forward slash. n=1, limits number of splits. expand=True returns as DataFrame, False returns data as series
crowdfunding_info_df[["category", "sub-category"]]=crowdfunding_info_df['category & sub-category'].str.split('/',n=1,expand=True)

crowdfunding_info_df.head(10)
```

```python
#find number of unique values in category 3 techniques:

# use for loop

count=0

unique=[]

for i in range(0,len(crowdfunding_info_df['category'])):
   
    if crowdfunding_info_df['category'][i] not in unique:
    
        unique.append(crowdfunding_info_df['category'][i])
    
        count += 1
    
print(count)

#use x = len()
x = len(crowdfunding_info_df['category'].unique())
x

#use .nunique()

print(crowdfunding_info_df['category'].nunique())
```

```python
#find number of unique values in sub-category 3 techniques:

cnt=0
hello=[]

for i in range(0,len(crowdfunding_info_df['sub-category'])):
    
    if crowdfunding_info_df['sub-category'][i] not in hello:
        
        hello.append(crowdfunding_info_df['sub-category'][i])
        
        cnt += 1
print(cnt)

# use x = len()

x = len(crowdfunding_info_df['sub-category'].unique())
x

# use print(.nunique())

print(crowdfunding_info_df['sub-category'].nunique())
```

```python
#create category and sub-category Dataframe
#first create separate varibles for each by using .unique()

category= crowdfunding_info_df['category'].unique()

subcategory = crowdfunding_info_df['sub-category'].unique()

print(category)

print()

print(subcategory)
```

```python
# to execute plan to create category and subcategory DF
# 1. create two NumPy arrays

import numpy as np

# create numpy arrays that have 1-10 for category and 1-25 for subcategory
#using np.arange(x,x)

category_ids=np.arange(1,10)

subcategory_ids=np.arange(1,25)

print(category_ids)

print()

print(subcategory_ids)
```

```python
# Use a list comprehension to add "cat0" to each category_id. 
cat_ids = ["cat0" + str(cat_id) for cat_id in category_ids]

# Use a list comprehension to add "scat0" to each subcategory_id. 
scat_ids = ["scat0" + str(scat_id) for scat_id in subcategory_ids ]

print(cat_ids)
print()
print(scat_ids)
```

```python
#create category Datafram using categoty and cat_ids using pd.DataFrame({'x':x,'x':x})

category_df=pd.DataFrame({'category_id' : cat_ids,
                         "category" : category})

category_df.head()

#create subcategory DataFrame using subcategory and scat_ids using pd.DataFrame({'x':x,'x':x})
subcategory_df = pd.DataFrame({'subcategory_id': scat_ids,
                              'sub-category' : subcategory})
subcategory_df.head()
```

```python
#save csv file

category_df.to_csv(r'Data/category.csv',index=False)

subcategory_df.to_csv(r'Data/subcategory.csv',index=False)
```

```python
#make copy of crowdfunding_info_df into campaign_df

campaign_df=crowdfunding_info_df.copy()

campaign_df.head(10)
```

```python
# #rename the blurb to description, launched_at to launch_date and deadline to end_date

campaign_df=campaign_df.rename(columns={'blurb':'description', 'launched_at':'launched_date','deadline':'end_date'})

campaign_df.head()

#change the goal and pledged columns from int64 to float

campaign_df[['goal','pledged']]=campaign_df[['goal','pledged']].astype(float)

campaign_df
```

```python
#convert the launched_date and end_date columns to international organization standardization(ISO) format
# 1. import "from datetime import date as dt"
# 2. use df['columns name']=pd.to_datetime(df['column name'],unit='s').dt.strftime('%Y-%m-%d')

# from datetime import date as dt

# campaign_df['launched_date']=pd.to_datetime(campaign_df['launched_date'], unit='s').dt.strftime('%Y-%m-%d')
# campaign_df['end_date']=pd.to_datetime(campaign_df['end_date'], unit='s').dt.strftime('%Y-%m-%d')

# #print

campaign_df.head()

```

```python
# Next, we want to assign each category and subcategory in the campaign_df DataFrame the unique 
# "category_id" and "subcategory_id" number from the category_df and subcategory_df DataFrame, respectively.

#use the merge fuction df1.merge(df2 on="column", how=left)

# Merge the campaign_df with the category_df on the "category" column and 
# the subcategory_df on the "subcategory" column.
campaign_merged_df = campaign_df.merge(category_df, on='category', how='left').merge(subcategory_df, on='sub-category', how='left')

campaign_merged_df.tail(10)
```

```python
# Drop unwanted columns.
campaign_cleaned = campaign_merged_df.drop(['staff_pick', 'spotlight', 'category & sub-category','category', 'sub-category'], axis=1)
campaign_cleaned.head()
```

```python
#show the contact_info_df DataFrame

contact_info_df.head()

# # Retrieve the data from the "contact_info" column.
# contact_info_list = contact_info_df.contact_info.to_list()
# contact_info_list

# Pop out the unique identification number using list slicing. 
# print(contact_info_list[0])
# print(contact_info_list[0][15:19])

# print([x[15:19] for x in contact_info_list])

# Extract the four-digit contact ID number and add it to a new column in the `campaign_cleaned` DataFrame.
campaign_cleaned["contact_id"] = [x[15:19] for x in contact_info_df["contact_info"].values]
campaign_cleaned.head()
```

```python
#changed data type of contact_id to int64 by usind df['column']=df['column'].apply(pd.to_numeric)

campaign_cleaned['contact_id']=campaign_cleaned['contact_id'].apply(pd.to_numeric)

campaign_cleaned.dtypes
```

```python
# rearrange columns in the data with df=df[['column1','column2','column3',etc..]]

campaign_cleaned=campaign_cleaned[["cf_id", "contact_id", "company_name", "description", "goal", "pledged", "outcome", "backers_count", "country", "currency", "launched_date", "end_date", "category_id","subcategory_id"]]

campaign_cleaned.head()

#save new Dataframe as campaign.csv using df.to_csv(), with encoding='utf8' and index=False
campaign_cleaned.to_csv(r'Data/campaign.csv',encoding='utf8',index=False)
```
